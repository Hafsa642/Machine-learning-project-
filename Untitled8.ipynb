```python
# 1. Load Libraries & Dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
# 2. Basic Dataset Overview
df.head()
df.info()
df.describe()
```

```python
# 3. Missing Values Handling
df.isnull().sum()
```

```python

df = df.dropna(thresh=int(df.shape[1]*0.7))

df = df.fillna(df.median(numeric_only=True))
df = df.fillna(df.mode().iloc[0])
```

```python
df = df.drop_duplicates()
```

```python
yes_no_cols = [c for c in df.columns if df[c].isin(["Yes","No"]).any()]

for col in yes_no_cols:
    df[col] = df[col].replace({"Yes": 1, "No": 0}).astype(int)
```

```python
plt.figure(figsize=(4,3))
df["Churn"].value_counts().plot(kind="bar")
plt.title("Churn Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
numeric_cols = df.select_dtypes("number").columns

for col in numeric_cols:
    plt.figure(figsize=(4,3))
    sns.histplot(df[col], kde=True)
    plt.title(col)
    plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.countplot(x="Contract", hue="Churn", data=df)
plt.title("Contract Type vs Churn")
plt.show()
```

```python
df.to_csv("churn_cleaned.csv", index=False)
```
